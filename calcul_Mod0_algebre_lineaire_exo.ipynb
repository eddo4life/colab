## 2. Produit Scalaire et Orthogonalité

```python
# Produit scalaire et angle entre vecteurs
def angle_between_vectors(u, v):
  """Calcule l'angle entre deux vecteurs en degrés"""
  # TODO: Calculer le produit scalaire u.v
  dot_product = np.dot(u, v)
  # TODO: Calculer les normes de u et v
  norm_u = np.linalg.norm(u)
  norm_v = np.linalg.norm(v)
  # TODO: Calculer le cosinus de l'angle
  cos_angle = dot_product / (norm_u*norm_v)
  # TODO: Convertir en degrés (utiliser arccos)
  angle_deg = np.degrees(np.arccos(cos_angle))
  return angle_deg
# Test
u = np.array([1, 0, 1])
v = np.array([1, 1, 1])
# u = np.array([1, 0])
# v = np.array([1, 1])
print(f"Angle entre {u} et {v}: {angle_between_vectors(u, v):.2f}° (Attendu: 45.00°)")
```

## 3. Normes de Vecteurs

```python
# Comparaison des différentes normes
v = np.array([3, 4, 5, 6])# Calcul des normes
# TODO: Calculer la norme L1 (Manhattan)
norm_1 = np.linalg.norm(v, 1)
# TODO: Calculer la norme L2 (Euclidienne)
norm_2 = np.linalg.norm(v, 2)
# TODO: Calculer la norme L-infini (Maximum)
norm_inf = np.linalg.norm(v, np.inf)
print(f"Vecteur v = {v}")
print(f"Norme L1: {norm_1}")
print(f"Norme L2: {norm_2}")
print(f"Norme L∞: {norm_inf}")
```

## 4. Opérations Matricielles

```python
# Opérations sur les matricesprint("OPÉRATIONS MATRICIELLES\n" + "="*60 + "\n")
# Définition des matrices# TODO: Définir la matrice A = [[1, 2, 5], [3, 4, 6], [1,2,3]]
A = np.array([[1, 2, 5], [3, 4, 6], [1,2,3]])
# TODO: Définir la matrice B = [[5, 6, 10], [7, 8, 1], [4,5,6]]
B = np.array([[5, 6, 10], [7, 8, 1], [4,5,6]])
print("Matrice A:")
print(A)
print("\nMatrice B:")
print(B)
# Addition
print("\n" + "-"*60)
# TODO: Calculer A + B
print("A + B =\n", A+B)
# Multiplication matricielle
print("\n" + "-"*60)
# TODO: Calculer le produit matriciel A @ B
print("A · B =\n", A @ B)
```

## 5. Matrice Inverse

```python
# Calcul de l'inverse d'une matrice
print("INVERSION DE MATRICE\n" + "="*60 + "\n")
A = np.array([[2, 1, 1],   [1,0,1], [4, 5, 3]])
print("Matrice A:")
print(A)
# Déterminant
# TODO: Calculer le déterminant de A
det_A = np.linalg.det(A)
print(f"\nDéterminant: det(A) = {det_A}")

if det_A != 0:
  # TODO: Calculer l'inverse de A
  A_inv = np.linalg.inv(A)
  print("\nInverse A⁻¹:")
  print(A_inv)
  # Vérification
  print("\nVérification A · A⁻¹ (doit être l'identité):")
  print(np.dot(A, A_inv))
```

## 6. Systèmes Linéaires

```python
# Résolution de systèmes linéaires
print("RÉSOLUTION DE SYSTÈME LINÉAIRE\n" + "="*60 + "\n")
print("Système:")
print("  2x + y -z = 5")
print("  x - 3y + 5z = -4")
print("  3x - 1y + 7z = 1")
print("\nForme matricielle: A·x = b\n")
# TODO: Définir la matrice des coefficients
A = np.array([[2,1,-1],[1,-3,5],[3,-1,7]])
# TODO: Définir le vecteur second membre
B = np.array([5,-4,1])
# TODO: Résoudre le système pour trouver x
x_sol = np.linalg.solve(A, B)
# print(f"Solution trouvée: x={x_sol[0]:.2f}, y={x_sol[1]:.2f}")
print(x_sol)
print(f"Solution trouvée: x={x_sol[0]:.2f}, y={x_sol[1]:.2f}, z={x_sol[2]:.2f}")
```

## 7. Application ML: Régression Linéaire

```python
# Régression linéaire avec algèbre linéaire
print("RÉGRESSION LINÉAIRE PAR MOINDRES CARRÉS\n" + "="*60 + "\n")
# Données synthétiques
np.random.seed(42)
n = 20
x_data = np.linspace(0, 10, n)
y_true = 2*x_data + 3
# Vraie relation
y_data = y_true + np.random.normal(0, 2, n)
# Avec bruit
# Matrice de design X (colonne de 1 et colonne de x)
# TODO: Créer la matrice X (indice: utiliser np.vstack et .T)
X = np.column_stack([np.ones(n), x_data])
# Résolution par équation normale:
# theta = (X.T X)^-1 X.T y
# TODO: Calculer theta
# theta = np.linalg.solve(X.T @ X, X.T @ y_data)
theta = np.linalg.inv(X.T @ X) @ X.T @ y_data
print(f"Paramètres estimés: Intercept={theta[0]:.2f}, Pente={theta[1]:.2f}")
print("Vrais paramètres: Intercept=3.00, Pente=2.00")
```
